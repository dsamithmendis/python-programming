```python
# ---------------------------------------------
# ğŸš´â€â™‚ï¸ BIKE RENTAL SYSTEM (OOP DEMONSTRATION)
# ---------------------------------------------

# --------------------------
# Task 1: Define Bicycle Class
# --------------------------

class Bicycle:
    """
    This class represents a Bicycle.
    It follows the principles of Object-Oriented Programming (OOP),
    particularly ENCAPSULATION â€” grouping related data (attributes)
    and behaviors (methods) together.
    """

    # The __init__ method is a special method called a CONSTRUCTOR.
    # It initializes the object's attributes when the object is created.
    def __init__(self, gear=1, speed=0):
        # 'self' refers to the specific object being created.
        # gear and speed are instance variables (unique to each object).
        self.gear = gear
        self.speed = speed

    # Method to change the gear â€” an example of an object's BEHAVIOR.
    def change_gear(self, new_gear):
        self.gear = new_gear
        print(f"âœ… Gear changed to {self.gear}")

    # Method to apply brakes â€” modifies the object's internal state.
    def apply_brakes(self):
        self.speed = 0
        print("ğŸ›‘ Brakes applied. The bike has stopped.")

    # __str__ is a special method (called a MAGIC or DUNDER method).
    # It defines what happens when you print the object.
    def __str__(self):
        return f"Bicycle(gear={self.gear}, speed={self.speed})"


# --------------------------
# Task 2: Create Bicycle Object and Test Methods
# --------------------------

# Creating a Bicycle object (an INSTANCE of the class)
# This shows INSTANTIATION â€” creating an object from a class blueprint.
print("\n--- Bicycle Object Creation ---")
myBike = Bicycle()  # Default gear=1, speed=0
print("ğŸš² Initial bike state:", myBike)

# Taking user input to change gear and speed
new_gear = int(input("Enter new gear for the bike: "))
new_speed = int(input("Enter new speed for the bike: "))

# Change gear and manually update speed
myBike.change_gear(new_gear)
myBike.speed = new_speed
print("ğŸš´ Updated bike state:", myBike)

# Apply brakes and print again
myBike.apply_brakes()
print("ğŸš´ After applying brakes:", myBike)


# --------------------------
# Task 3: Define Customer Class
# --------------------------

class Customer:
    """
    Represents a customer who rents a bike.
    Demonstrates COMPOSITION â€” this class includes an object
    of another class (Bicycle) as an attribute.
    """

    # Constructor to initialize the customer's attributes
    def __init__(self, customer_id, name, rented_bike=None):
        self.customer_id = customer_id
        self.name = name
        self.rented_bike = rented_bike  # Can hold a Bicycle object

    # Method to rent a bike (assign a Bicycle object to the customer)
    def rent_bike(self, bike):
        self.rented_bike = bike
        print(f"ğŸš² {self.name} has rented the bike.")

    # Method to return the rented bike
    def return_bike(self):
        if self.rented_bike:
            print(f"ğŸ” {self.name} has returned the bike.")
            self.rented_bike = None
        else:
            print("âš ï¸ No bike to return.")

    # String representation of the Customer
    def __str__(self):
        if self.rented_bike:
            return f"Customer(ID: {self.customer_id}, Name: {self.name}, Rented: {self.rented_bike})"
        else:
            return f"Customer(ID: {self.customer_id}, Name: {self.name}, Rented: None)"


# --------------------------
# Task 4: Simulate Bike Rental Process
# --------------------------

print("\n--- Bike Rental Simulation ---")

# Create a Customer object named 'john'
john = Customer("C123", "John Doe")

# Simulate John renting the previously created bike
john.rent_bike(myBike)
print("ğŸ‘¤ Customer Info:", john)
print("ğŸš² Bike Info:", myBike)

# Simulate John returning the bike
john.return_bike()
print("ğŸ‘¤ Updated Customer Info:", john)
print("ğŸš² Bike Info After Return:", myBike)


# --------------------------
# Activities
# --------------------------

# Activity 1: Define Student class
print("\n--- Activity 1: Student Class ---")
class Student:
    """
    Represents a student with ID, last name, and course.
    Demonstrates basic CLASS and OBJECT usage.
    """
    def __init__(self, studentID, lastName, courseID):
        self.studentID = studentID
        self.lastName = lastName
        self.courseID = courseID

    def __str__(self):
        return f"Student(ID: {self.studentID}, Name: {self.lastName}, Course: {self.courseID})"


# Create and print a Student object
student1 = Student("S001", "Mendis", "CE202")
print(student1)


# Activity 2: Reuse the Bicycle class (already done above)
print("\n--- Activity 2: Change Bicycle Gear ---")
bike2 = Bicycle()
print("Before gear change:", bike2)
bike2.change_gear(2)
print("After gear change:", bike2)


# Activity 3: Demonstrate __str__ method (already shown)
print("\n--- Activity 3: Bicycle __str__ Method ---")
print("Bicycle current state:", bike2)
```
